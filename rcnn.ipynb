```python
import cv2
import pixellib
from pixellib.instance import instance_segmentation
```

```python
seg_model = instance_segmentation()
seg_model.load_model('mask_rcnn_coco.h5')
```

```python
capture = cv2.VideoCapture(0)
capture.set(3, 1200)  # 3 for width
capture.set(4, 900)  # 4 for height
capture.set(10, 100)  # 10 for brightness

while True:
    ret, frame = capture.read()
    frame = cv2.flip(frame, 1)
    
    seg = seg_model.segmentFrame(frame, show_bboxes=True)
    img = seg[1]
    
    cv2.imshow('Out', img)

    if cv2.waitKey(1) == 27:  # ESC key to break
        break

cv2.destroyAllWindows()
capture.release()
```
